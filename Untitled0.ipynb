```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('sales_dataset.csv')
#  1 Histogram of TotalSaless
sns.histplot(df['TotalSales'], bins=10, kde=True)
plt.title('Histogram of Total Sales')
plt.show()
```

```python
# 2 Box Plot of Price

sns.boxplot(x='Category', y='Price', data=df)
plt.title('Box Plot of Price by Category')
plt.show()
```

```python
# 3 Line Plot of Total Sales over Date
df['Date'] = pd.to_datetime(df['Date'])
df_sorted = df.sort_values('Date')
sns.lineplot(x='Date', y='TotalSales', data=df_sorted, marker='o')
plt.title('Line Plot of Total Sales Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4 Bar Plot of Quantity by Product
sns.barplot(x='Product', y='Quantity', data=df)
plt.title('Bar Plot of Quantity by Product')
plt.xticks(rotation=90)
plt.show()
```

```python
# 5 Scatter Plot: Price vs TotalSales
sns.scatterplot(x='Price', y='TotalSales', hue='Category', data=df, s=100)
plt.title('Scatter Plot of Price vs Total Sales')
plt.show()
```

```python
# 6 Relplot (Relationship Plot)
sns.relplot(x='Quantity', y='TotalSales', hue='Category', kind='scatter', data=df, height=6, aspect=1.5)
plt.title('Relplot of Quantity vs Total Sales')
plt.show()
```

```python
# 7 Joint Plot (Price vs TotalSales)
sns.jointplot(x='Price', y='TotalSales', data=df, kind='hex')
plt.show()
```

```python
# 8 PairGrid (Multivariate Relationship)
g = sns.PairGrid(df, hue='Category', height=3)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
```

```python
# 9 Plot Heatmap
numerical_df = df[['Quantity', 'Price', 'TotalSales']]
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Feature Correlations')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
"Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
"Sales": [250, 300, 400, 350, 500, 600],
"Profit": [70, 90, 120, 100, 150, 200]
}
df = pd.DataFrame(data)

#Print data frame
print(df)

#Print line graph
df.plot(x="Month", y="Sales", kind="line", marker="o",title="Monthly Sales")
plt.ylabel("Sales in $")
plt.show()

#Print bar graph
df.plot(x="Month", y="Profit", kind="bar", title="Monthly Profit", color="orange")
plt.ylabel("Profit in $")
plt.show()

#Print line graph for sales vs profit
df.plot(x="Month", y=["Sales", "Profit"], kind="line", marker="o")
plt.title("Sales vs Profit Trend")
plt.show()

#Scatter plot
df.plot.scatter(x="Sales", y="Profit", title="Sales vs Profit", color="orange")
plt.show()
```
